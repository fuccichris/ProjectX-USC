```
def generator_date(date, session):
    df = pd.DataFrame()
    for period in range(1, 49):
        url = "https://api.bmreports.com/BMRS/B1610/v2?APIKey=c1e8s775q9a354o&SettlementDate={}&Period={}&ServiceType=csv".format(date, period)
        download = session.get(url.format(date, period)) # download csv
        decoded_content = download.content.decode('utf-8') # decode bytes response
        decoded_content = decoded_content[decoded_content.find("\n")+1:] # remove first line

        # load df
        decoded_content = io.StringIO(decoded_content)
        df = pd.concat((df, pd.read_csv(decoded_content, sep=",")))
    return df
```

```
years = range(2018, 2021)
df = pd.DataFrame()
with requests.Session() as s:
    for year in years:
        for date in tqdm(pd.date_range(start=f"{year}-01-01", end=f"{year}-12-31", freq="D")):
            try:
                df = pd.concat((df, generator_date(date, s)))
            except:
                print("Error in %s " % date)
        df.to_csv("generator_data_3.csv")
```
